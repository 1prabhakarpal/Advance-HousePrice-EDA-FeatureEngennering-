# Missing Values

```python
# Missing Values in categorical variable
feature_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtype =='O']

for feature in feature_with_nan:
    print('{} : {}% missing values'.format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
# Replace missing values with new values
def replace_cat_feature(df,feature_nan):
    data = df.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

df = replace_cat_feature(df,feature_with_nan)
df[feature_with_nan].isnull().sum()
```

```python
df.head()
```

```python
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes != 'O']
for feature in numerical_with_nan:
    print('{} : {}% missing values'.format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
for feature in numerical_with_nan:
    median = df[feature].median()
    
    df[feature+'nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median,inplace = True)

df[numerical_with_nan].isnull().sum()
```

```python
df.head(50)
```

```python
# Temporal Variable 
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature] = df['YrSold'] - df[feature]
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

#  Numerical Variables
since the numerical features are skewed we will perform log normal distribution

```python
df.head()
```

```python
num_feature = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_feature:
    df[feature] = np.log(df[feature])
```

```python
df.head()
```

# Handeling rare categorical feature
remove categorical that are present less then 1% of the observation

```python
cat_feature = [feature for feature in df.columns if df[feature].dtypes == 'O']
```

```python
cat_feature
```

```python
for feature in cat_feature:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[feature] = np.where(df[feature].isin(temp_df),df[feature],'Rare_Var')
```

```python
df.head()
```

```python
for feature in cat_feature:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head()
```

```python
feature_scale = [feature for feature in df.columns if feature not in  ['Id','SalePrice']]
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[feature_scale])
```

```python
scaler.transform(df[feature_scale])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([df[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(df[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
